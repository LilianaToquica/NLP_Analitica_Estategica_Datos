```python
import urllib.request
import bs4 as bs # Beautiful Soup

link = "https://es.wikipedia.org/wiki/Carl_Sagan"

# Descargar página web en Python
request = urllib.request.urlopen(link)
source = request.read() 
request.close()

soup = bs.BeautifulSoup(source,'html.parser') 

for e in soup.find_all('p')[0].find_all('a'):
    print(e)
    print("Contenido: ")
    print(e.contents)
    print(f"Título: ")
    print(e.get('title'))
    print()
```
