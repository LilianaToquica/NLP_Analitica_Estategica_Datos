# Punto 1:

- Descomprimir el archivo `.zip` de los poemas

```python
import os
import shutil
from zipfile import ZipFile 
```

```python
direccion = 'archivos/poemas.zip'
```

```python
# Descomprimir el archivo ZIP
with ZipFile(direccion) as archivo: 
    archivo.extractall('archivos/') 
    shutil.rmtree('archivos/__MACOSX')
```

- Leer cada uno de sus archivos

```python
dict_respuestas = {}

for raiz, dirs, archivos in os.walk('archivos/poemas'):
        for archivo in archivos:
            print(f"Nombre del archivo: {archivo}")
            
            # Crear ruta
            ruta = f"archivos/poemas/{archivo}"
            print(f"Ruta del archivo: {ruta}\n")
            
            # Leer archivo
            with open(ruta) as poema: 
                data = poema.read()
            
            print("POEMA")
            print(data)
            
            # Contar número de palabras
            num_palabras = len(data.split())
            print(f"\n>>>>> Número de palabras: {num_palabras}")
            
            # Agregar a diccionario
            dict_respuestas[archivo] = num_palabras
            
            print("*"*100)
            print()
```

- Responder: ¿Cuál archivo tiene el mayor número de palabras?

```python
dict_respuestas
```

```python
dict_respuestas_ordenado = sorted(dict_respuestas.items(), key=lambda registro: registro[1], reverse=True)
dict_respuestas_ordenado
```

```python
print(f"El archivo con más palabras es {dict_respuestas_ordenado[0][0]} con {dict_respuestas_ordenado[0][1]} palabras.")
```

# Punto 2:

- Hacer Web Scraping de 10 biografías en Wikipedia (en búcle)
- Obtener el **encabezado** de cada biografía
- Obtener todos los **contenidos** y **etiquetas de título** asociados a los **links** del primer párrafo

<img src='img/carlsagan.png'>

```python
import bs4 as bs 
import urllib.request
```

Lista de las 10 biografias

```python
mi_lista = ["https://es.wikipedia.org/wiki/Carl_Sagan",
            "https://es.wikipedia.org/wiki/Carl_Gustav_Jung",
            "https://es.wikipedia.org/wiki/Carl_Friedrich_Gauss",
            "https://es.wikipedia.org/wiki/Carlos_Vives",
            "https://es.wikipedia.org/wiki/Carlos_Valderrama",
            "https://es.wikipedia.org/wiki/Carlos_de_Gales",
            "https://es.wikipedia.org/wiki/Carlos_Calero",
            "https://es.wikipedia.org/wiki/Carlos_Santana",
            "https://en.wikipedia.org/wiki/Carlos_Slim",
            "https://es.wikipedia.org/wiki/Carlos_Pena,_Jr."]
```

```python
for i, enlace in enumerate(mi_lista):
    print(f"Enlace #{i+1:<2}: {enlace}")
    
    # Obtener página web en Python
    request = urllib.request.urlopen(enlace)
    source = request.read() 
    request.close()     
    soup = bs.BeautifulSoup(source,'html.parser') 
    
    # Encabezado
    encabezado = soup.find(id='firstHeading').contents[0]
    print(f"Encabezado: {encabezado}")
    
    # Primer párrafo
    parrafo = soup.find('p')
    
    # Iterar sobre los link del primer párrafo
    print("\n>>>> LINKS")
    for link in parrafo.find_all('a', href=True):
        print(f"Etiqueta: {link}")
        print(f"Título: {link.get('title')}")
        print(f"Contenido: {link.contents}")
        print()
    
    print("*"*100)
    print()
```
